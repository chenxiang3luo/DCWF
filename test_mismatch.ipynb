```python
import numpy as np
import torch
import torch as t
```

```python
import re

def clean_url(url):
    pattern = r"(?:https?://)?(?:www\.)?([^/?]+)"
    match = re.match(pattern, url)
    domain = match.group(1)
    return domain

# 示例用法
url = "https://www.example.com/path/to/page"
cleaned_url = clean_url(url)
print(cleaned_url)
```

```python
knn_webs = []
with open ('./knnlists.txt') as f:
    for line in f.readlines():
        knn_webs.append(clean_url(line.strip()))
```

```python
dataset = np.load('/root/datasets/FSCIL/tor_900w_2500tr.npz', allow_pickle = True)
```

```python
common = set(dataset['labels']).intersection(set(knn_webs))
```

```python
lab_900 = dataset['labels']
data_900 = dataset['data']
AWF_data_comm = []
AWF_labels_comm = []
for cl in common:
    inds = np.where(lab_900 == cl)[0]
    inds = inds.reshape(-1)
    sample = np.random.permutation(inds.shape[0])
    sample = inds[sample][:90]
    AWF_data_comm.extend(data_900[sample])
    AWF_labels_comm.extend(lab_900[sample])
```

```python
np.savez('./common_AWF',data = np.array(AWF_data_comm),labels = np.array(AWF_labels_comm))
```

```python
count = []
cls = []
c = 0
with open ('./knnlists.txt') as f:
    for line in f.readlines():
        for cl in common:
            if(cl in line):
                if(cl=='t.co' and not line.startswith('t.co')):
                    continue
                count.append(c)
                cls.append(cl.strip())
        c += 1
```

```python
knn = np.load('/root/datasets/FSCIL/KNN.npz',allow_pickle=True)
lab_knn = knn['labels']
data_knn = knn['data']
knn_data_comm = []
knn_labels_comm = []
for cc,cl in zip(count,cls):
    inds = np.where(lab_knn == str(cc))[0]
    inds = inds.reshape(-1)
    sample = np.random.permutation(inds.shape[0])
    sample = inds[sample][:90]
    knn_data_comm.extend(data_knn[sample])
    print(lab_knn[sample])
    ll = [cl for i in lab_knn[sample]]
    knn_labels_comm.extend(ll)
```

```python
np.savez('./common_KNN',data = np.array(knn_data_comm),labels = np.array(knn_labels_comm))
```
